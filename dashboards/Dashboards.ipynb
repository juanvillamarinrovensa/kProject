## KPIs

**KPIS definition**
1. define goals
2. per goal:
    + how to be measured
    + how often to be measured





### measurements and types

there is two types of measurements ***[6]***: **Operational** and **Strategic impact**

1. Operational
    + **Input:** resources consumed in processes that produce outputs
    + **process:** measures of efficiency, and quality to produce
    + **output:** result measures of completed work      
2. Strategic impact
    + **intermediate outcomes:** e.g. customer brand awareness (as a result of marketing  or communication)
    + **end outcomes:** e.g sales

and also two moments to be measured: **lagging** (past) and **leading** (future):

1. **Lagging KPIs:** -What has happpened?
    + *Where you business has been* 
    + Outcomes (output metrics)
    + High level metrics e.g. **sales revenue**
2. **Leading KPIs:** -Where are we going?
    + *Where your business is going*
    + Progress going how (input metrics)
    + faster feedback e.g website tracking

### stakeholders and design 

***a dashboard is viewed by people with different roles***
+ rarely do two people have the same exact goals
+ rarely do two people require the exact same information

therefore, understanding stakeholder needs will increase dashboard effectiveness

**types of stakeholders**



***extracting info from stakeholders***

what to ask..
+ ***why do you need a dashboard? what do you hope to achieve?*** - *i want to monitor financial health*
+ ***what are the essential KPIs needed to make the dashboard effective?*** - for an HR manager, employee turnover
+ ***where and how will the dashboard be used?*** - how frequent is the dashboard going to be used, only for a meeting, to stand in front of the factory door, etc. 
+ ***what are the company`s strategic goals?*** -  e.g. to increase financial revenue for the upcoming 2 years

### glossary 

1. **Data deluge:** It refers to the vast and rapidly increasing amount of data that is being created by digital technologies and devices, such as sensors, social media, and other sources. The term "data deluge" is used to emphasize the challenges of managing, analyzing, and making sense of this massive amount of data, which can easily overwhelm individuals and organizations. In order to cope with the data deluge, advanced data management and analysis techniques are required, including artificial intelligence, machine learning, and data visualization tools.

2. **Data lake:** A centralized repository for storing large amounts of structured and unstructured data at any scale. Unlike traditional data warehouses, which are designed for structured data, a data lake is designed to store all types of data, including raw and unprocessed data, as well as semi-structured and unstructured data. Data lakes typically use a schema-on-read approach, which means that data is not structured until it is read or queried, allowing for greater flexibility in data analysis. However, this also means that data quality and governance can be more challenging in a data  lake environment, and proper data management processes and tools are necessary to ensure data accuracy, consistency, and security. 

3. **Dashboard:** A visual display of the most important **information needed** to achieve one or more objectives, arranged on a single screen, to be monitored at a glance.
